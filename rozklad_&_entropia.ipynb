```python
import matplotlib.pyplot as plt
from scipy.stats import entropy
import pandas as pd
import numpy as np
```

```python
frames = np.loadtxt("numbers.txt", dtype=np.uint8)

plt.figure(figsize=[16,10])
plt.title("Empiryczny rozkład zmiennych losowych po postprocessingu", fontsize=15)
plt.xlabel("Wartość (x_i)", fontsize=15)
plt.ylabel("Częstość występowania (p_i)", fontsize=15)
plt.hist(frames, bins=256,density=True)
plt.show()
```

```python
frames = np.loadtxt("numbers.txt",dtype=np.uint8)

def calc_entropy(nums):
     p, _ = np.histogram(nums, bins=256, range=(0, 255))
     entropy_val = entropy(p, base=2)
     return entropy_val

calc_entropy(frames)
```
